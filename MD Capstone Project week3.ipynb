```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

```python
wikipedia_postalcode='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

### Using BeautifulSoup package and html parser to open URL and save the data into String

```python
ps_url = urllib.request.urlopen(wikipedia_postalcode)
ps_code_data = BeautifulSoup(ps_url,'html.parser')
print(ps_code_data.prettify)
```

```python
right_table=ps_code_data.find('table', class_='wikitable sortable')
```

#### Loading of data into array as column wise from html tag

```python
A=[]
B=[]
C=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states = row.findAll('th') #To store second column data        
            
    if len(cells)==3: #Only extract table body not heading            
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

#### loading the list data into data frame

```python
#To convert list to data frame
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighbourhood']=C
df.shape
```

#### filtering out the 'Not assigned' value from column 'Borough' and print the shape

```python
actual = df['Borough'] != 'Not assigned'
df_final = df[actual]
print(df_final.shape)
```

#### grouping of values and Concatenating the Neighbourhood  

```python
df_final1 = (df_final.groupby(['Postal Code','Borough'])['Neighbourhood'].apply(lambda x: ' , '.join(set(x.dropna()))).reset_index())
```

#### Identifying & assigning the Borough value to the Neighbourhood column

```python
df_final1['Neighbourhood'][df_final1.Neighbourhood == 'Not assigned\n'] = df_final1['Borough'][df_final1.Neighbourhood == 'Not assigned\n']
df_final1
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```
